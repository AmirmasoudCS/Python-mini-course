<p>A loop is a block of code that runs again and again until a condition stops it.</p>
<ol>
    <li> for loop → iterate over a sequence</li>
    <li> while loop → run while the condition is True</li>
</ol>

<h2>The <i>for</i> loop</h2>

```python
for i in range(0,6): # first-inclusive, last-exclusive
    print(i)
```

```python
fruits = ["Apple", "Banana", "Mango"]
for index in range(len(fruits)):
    print(fruits[index])
```

```python
for i in fruits:
    print(i , end=" ")
```

```python
for i in range(10):
    print(i)
```

```python
for i in range(24,100,2):
    print(i, end=",")
```

<h2>The <i>while</i> loop</h2>

```python
i = 0
while i < 5:
    print(i)
    i += 1
```

<h2>Loop control statements</h2>

<h3>break</h3>

```python
students = ["Amir", "Masoud", "Harry", "Kasra", "Arsham"] # Find Harry!
for student in students:
    if student == "Harry":
        print("I found you Harry!")
        break;
    else:
        print("You are not Harry, you are", student)
```

```python
students = ["Neville Longbottom", "Drako Malfoy", "Harry Potter", "Ron Weasley", "Hermione Granger"] # Don't mention Drako...
for student in students:
    if student == "Drako Malfoy":
        continue
    else:
        print(student, "is here")
```

```python
attendees = ["Minerva McGonagall", "Hermione Granger", "Severus Snape", "Tom Marvolo Riddle","Remus Lupin", "Luna Lovegood", "Fred Weasley"] # Say Hello to professors
students = ["Harry Potter", "Hermione Granger", "Ron Weasley", "Luna Lovegood", "Cedric Diggory", "Cho Chang"]
professors = ["Albus Dumbledore", "Minerva McGonagall", "Severus Snape", "Remus Lupin", "Alastar Mad-Eye Moody"]
```

```python
for attendee in attendees:
    if attendee in students:
        continue
    elif attendee in professors:
        print("Hello professor", attendee)
    elif attendee not in (professors,students):
        print("Who are you", attendee, "?")
        break
```
