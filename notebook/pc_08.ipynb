<h2>What is a <i>Function</i>?</h2>
<div>A function is a block of code that performs a specific task.</div>
<div>Instead of writing the same code again and again, we can put it inside a function and reuse it.</div>

```python
def greet():
    print("Hello")
```

```python
greet()
```

<h2>Why Do We Use Functions?</h2>
<div>
    Functions help us:
    <ul>
        <li> avoid repeating code</li>
        <li> make code cleaner</li>
        <li> organize programs better</li>
        <li> debug mroe easily</li>
        <li> reuse logic many times</li>
    </ul>
    Functions are like mini machines.<br>
    You give them input, they do some work, and sometimes they give back output.
</div>

```python
def greet2(name):
    print("Hello", name)
greet2(2)
```

```python
def add(a, b):
    print(a + b)
```

```python
add(2, 3)
```

```python
def add2(a, b):
    return a + b
```

```python
result = add2(2, 3)
```

```python
result
```

```python
summ = add(2, 3)
```

```python
summ # None
```

<h2>Higher-Order Functions</h2>
<div>
    <p>In Python, functions are treated like <b>normal values</b>.</p>
    <p>That means a function can be <b>passed</b> into another function.</p>
</div>

```python
def greet():
    print("Hello")
def run_function(f):
    f()
```

```python
greet()
```

```python
run_function(greet)
```

```python
# build a simple calculator
```

```python
def add(a, b):
    return a+b
def multiply(a, b):
    return a*b
def calculate(function, a, b):
    return function(a, b)
```

```python
operation = input("What operation do you want to do? (multiply,add)")
opr1 = int(input("Enter first operand."))
opr2 = int(input("Enter second operand."))
```

```python
if operation == "multiply":
    result = calculate(multiply, opr1, opr2)
elif operation == "add":
    result = calculate(add, opr1, opr2)
result
```

<h2>Explicit Type Hints</h2>
<div>
    <p>Python does not require types, but we can add them as hints.<br>
    These are called type annotations or type hints.</p>
    <p>They help:</p>
        <ul>
            <li> readability</li>
            <li> documentation</li>
            <li> IDE suggestions</li>
            <li> static type checkers</li>
        </ul>
</div>

```python
# def function( parameter : type) -> return_type:
```

```python
def greet(name : str):
    print("Hello",name)
```

```python
greet("masoud")
```

```python
greet(1)
```

```python
greet(True)
```

```python
def multiply(a : float, b : float) -> float:
    return a*b
```

```python
result = multiply(2.3, 3.4)
result
```

<b>Type Hints do <i>NOT</i> enforce anything</b>

```python
def add(a : int, b : int) -> int:
    return a + b
def calculate(function, x : int, y : int) -> int:
    return function(x, y)
```

```python
calculate(add, 2.3, 3)
```

```python
from typing import Callable
```

```python
def calculate(function : Callable[[str,bool],float], x : int, y : int) -> int:
    return function(x,y)
```

```python
calculate(add, 2.3 , 3)
```
