<h2>Reading a File ("r")</h2>

```python
filename = "../students.txt"
file = open(filename, "r")
```

```python
file.seek(0)
for line in file:
    print(line)
```

```python
# each line
file.seek(0)
for line in file:
    fname,lname = line.strip().split()
    print(lname)
```

```python
file.seek(0)
myWholeFile = file.read()
```

```python
file.close()
```

<h3>Using <i>with</i> </h3>

```python
with open("../students.txt", "r") as f:
    content = f.read()
```

```python
content
```

```python
with open("../students.txt", "r") as file:
    lines = file.readlines()
```

```python
lines
```

<h2>Writing Files ("w")</h2>

```python
# make a file and write to it then read
with open("professors.txt","w") as f:
    f.write("Albus Dumbledore")
```

<h2>Appending Files ("a")</h2>

```python
# append previously made file
with open("professors.txt", "a") as f:
    f.write("\nSeverus Snape")
```

```python
with open("professors.txt", "r") as f:
    content = f.read()
content
```
