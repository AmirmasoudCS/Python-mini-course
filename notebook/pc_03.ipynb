<h2>Special Parameters</h2>

<div>
    <ul>
        <li> sep</li>
        <li> end</li>
    </ul>
</div>

```python
year = 2026
month = 5
day = 6
# Goal = 2026-5-6
print(f"{year}-{month}-{day}")
```

```python
print(year, month, day, sep="-")
```

```python
print(year, month, day, sep="ABCD")
```

```python
print("Hello")
print("World!")
# Goal = Hello World!
```

```python
print("Hello ", end="")
print("World!")
```
