<h2>Different Ways to Import Modules</h2>

<h3>A) Full Import</h3>

```python
import math
print(math.pi)
```

<h3>B) Import with alias</h3>

```python
import math as m
print(m.pi)
```

<h3>C) Import specific functions</h3>

```python
from math import sqrt, pi
print(sqrt(49) , pi)
```

<h3>D) Import Everything</h3>

```python
from math import *
math.pi
```
