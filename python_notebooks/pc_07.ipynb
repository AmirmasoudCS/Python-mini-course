<h2>Changing list items</h2>
<div>Lists are <b>mutable</b></div>

```python
grades = [-0.2, 15, True, 14, 15.6, "Amirmasoud", 19, ["Hello", "World", "1", 11.2222]] # remove non-grades
```

```python
grades
```

```python
for item in grades:
    if isinstance(item,(float, int)) and not isinstance(item, bool):
        continue
    else:
        grades.remove(item)
grades
```

```python
grades = [item for item in grades if isinstance(item,(int,float)) and not isinstance(item,bool) and item > 0]
grades
```

<h2>Dictionaries</h2>

<h3>What is a Dictionary?</h3>
<div>A dictionary stores data in key-value pairs.</div>

```python
professor = {
    "fname" : "Albus",
    "lname" : "Dumbledore",
    "role" : "Hogwarst Head-master"
}
```

```python
professor
```

```python
professor['fname']
```

```python
 #professor[0]
```

```python
professor['age'] = 83
```

```python
professor
```

```python
del professor['age']
```

```python
professor
```

```python
student = {
    'first_name' : "Harry",
    "last_name" : "Potter",
    "age" : 14,
    "favorite_class" : "Defence Against the Dark Magic",
    "favorite_professor" : "Remus Lupin"
}
```

```python
for key in student:
    print(key, "is", student[key])
```

```python
for key, value in student.items():
    print(key, "is", value)
```
