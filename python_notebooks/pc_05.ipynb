<h2>The <i>if</i> statement</h2>

```python
age = int(input("Enter your age:"))
```

```python
if age >= 18 :
    print("Adult")
```

<h2>The <i>else</i> statement</h2>

```python
if age >= 18:
    print("Adult")
else:
    print("Minor")
```

```python
mark = int(input("Enter your mark:"))
```

```python
if mark < 70:
    print("D")
else:
    if mark > 90:
        print("A")
    elif mark < 90 and mark > 80:
        print("B")
    else:
        print("C")
```

<h2>The <i>elif</i> statement</h2>

```python
if mark >= 90 :
    print("A")
elif mark > 80 and mark < 90:
    print("B")
else:
    print("C")
```

```python
day = input("What day is it today?")
```

```python
if day == "Saturday" or day == "Sunday":
    print("Have a nice weekend!")
else:
    print("Go back to WORK!")
```

```python
if day in ("Saturday", "Sunday"):
    print("Have a nice weekend!")
else:
    print("Go back to work!")
```
